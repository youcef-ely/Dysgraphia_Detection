# Load the data

```python
data = dm.read_data('dysgraphic_people_samples.xlsx')
data
```

```python
poly_data = dp.PolynomialFeatures_labeled(data.drop(['Date', 'ID'], axis = 1), 2)
poly_data.head()
```

```python
features = ['BHK_raw_quality_score**1 x median_Freq_speed**1',
 'BHK_raw_quality_score**1 x Age**1',
 'BHK_raw_speed_score**1 x mean_Pressure**1',
 'median_Freq_speed**1 x bandwidth_tilt_x**1']
```

```python
samples = poly_data[features]
samples[['Date', 'ID']] = data[['Date', 'ID']]
samples.head()
```

```python
evolution_data = data.copy()
```

```python
for id_ in data.ID.unique():
    predictions, probs = loaded_model.predict(samples[samples['ID'] == id_][features]), loaded_model.predict_proba(samples[samples['ID'] == id_][features])
    evolution_data.loc[evolution_data['ID'] == id_, 'Dysgraphia'] = predictions
    evolution_data.loc[evolution_data['ID'] == id_, 'Dysgraphia_Proba'] = probs[:, 1]
```

```python
evolution_data
```

```python
id_ = 280
```

```python
person = evolution_data[evolution_data['ID'] == id_]
```

```python
fig, ax1 = plt.subplots(nrows=1, ncols=1, figsize = (25, 8))


ax1.plot(person['Date'], person['Dysgraphia_Proba'], '-o', label = 'Evolution')
ax1.set_xlabel('Dates')
ax1.set_ylabel('Probability')

fig.autofmt_xdate()
plt.legend()
plt.show()
```
