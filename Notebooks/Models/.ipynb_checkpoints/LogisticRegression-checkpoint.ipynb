# Original Data 

```python
data = dm.read_data('reduced_data.xlsx')
data.head()
```

```python
X = data.drop(['ID', 'Dysgraphia'], axis = 1)
X.head()
```

```python
y = data['Dysgraphia']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0, shuffle=True, 
                                                    stratify = y)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
tst = Test(LogisticRegression(max_iter = 1000, random_state = 0))
tst.cross_validation_scores(X_train, y_train)
```

```python
model = tst.train_test_report(X_train, y_train, X_test, y_test)
```

# Polynomial Data

```python
poly_data = dp.PolynomialFeatures_labeled(X, 2).drop(['Constant Term'], axis = 1)
poly_data.head()
```

```python
from sklearn.feature_selection import SequentialFeatureSelector
```

```python
poly_X_train, poly_X_test, poly_y_train, poly_y_test = train_test_split(poly_data, y, test_size = 0.3, random_state = 0, shuffle=True, 
                                                        stratify = y)
poly_scaler = StandardScaler()
poly_X_train = poly_scaler.fit_transform(poly_X_train)
poly_X_test = poly_scaler.transform(poly_X_test)
```

```python
poly_X_train = pd.DataFrame(poly_X_train, columns = poly_data.columns)
poly_X_test = pd.DataFrame(poly_X_test, columns = poly_data.columns)
```

```python
print('Train size: {}, Test size: {}'.format(poly_X_train.shape[0], poly_X_test.shape[0]))
```

```python
selector = SequentialFeatureSelector(LogisticRegression(max_iter = 1000, random_state = 0),
                                    cv = 5, scoring = 'recall', n_features_to_select = 'auto',
                                    direction = 'forward', tol = 0.00000001)
```

```python
selector.fit(poly_X_train, poly_y_train)
```

```python
new_features = list(selector.get_feature_names_out())
```

```python
new_features
```

### 'median_Freq_speed**1', 'BHK_raw_quality_score**1 x BHK_raw_speed_score**1',  'BHK_raw_quality_score**1 x Age**1'

```python
sfs_test = Test(model = LogisticRegression(random_state = 0, max_iter = 1000), n_folds = 5)
scores = sfs_test.cross_validation_scores(poly_X_train[new_features], poly_y_train)
scores
```

```python
sfs_test.train_test_report(poly_X_train[new_features], y_train, poly_X_test[new_features], y_test)
```

# Hyper-Parameters Optimization

```python
from genetic_algorithm import GeneticAlgorithm
from sklearn.model_selection import RandomizedSearchCV
```

```python
Grid_params = [
    {'solver' : ['saga'],
      'penalty' : ['elasticnet', 'l1', 'l2', None],
      'max_iter' : [50,100,200,500,1000,2500],
      'C' : [0.001, 0.01, 0.1, 1, 10, 100, 1000],
      'random_state': [0]},
    {'solver' : ['newton-cg', 'lbfgs'],
      'penalty' : ['l2',None],
      'max_iter' : [50, 100, 200, 500, 1000, 2500],
      'C' : [0.001, 0.01, 0.1, 1, 10, 100, 1000],
      'random_state': [0]}
    ]
```

```python
grid_optmizer = RandomizedSearchCV(LogisticRegression(),
                               param_distributions = Grid_params, cv = 5, scoring = 'recall', n_iter = 100, verbose = 0)
```

```python
grid_optmizer.fit(poly_X_train[new_features], y_train)
```

```python
params = grid_optmizer.best_params_
params
```

```python
sfs_test = Test(model = LogisticRegression(**params), n_folds = 5)
scores = sfs_test.cross_validation_scores(poly_X_train[new_features], poly_y_train)
scores
```

```python
model = sfs_test.train_test_report(poly_X_train[new_features], y_train, poly_X_test[new_features], y_test)
```

```python
sfs_test.precision_recall_curve(poly_X_train[new_features], poly_y_train)
```

```python
def final_model(model, X, threshold = 0):
    return model.decision_function(X) > threshold
```

```python
predictions = final_model(grid_optmizer.best_estimator_, poly_X_test[new_features], -1)
```

```python
from sklearn.metrics import accuracy_score, recall_score, f1_score, precision_score
```

```python
print('Accuracy: {}, Recall: {}, Precision: {}, F1: {}'.format(accuracy_score(y_test, predictions), recall_score(y_test, predictions), precision_score(y_test, predictions), f1_score(y_test, predictions)))
```
