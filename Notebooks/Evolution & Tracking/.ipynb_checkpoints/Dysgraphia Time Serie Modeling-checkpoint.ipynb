```python
import pandas as pd
```

```python
original_data = pd.read_excel('/content/reduced_data.xlsx')
original_data.head()
```

```python
original_data.head()
```

```python
features = ['BHK_raw_quality_score', 'BHK_raw_speed_score', 'median_Freq_speed', 'Space_Between_Words']
```

```python
dys_df = original_data[original_data['Dysgraphia'] == 1][['ID'] + features + ['Age']]
normal_df = original_data[original_data['Dysgraphia'] == 0][['ID'] + features + ['Age']]
dys_df
```

```python
dysgraphia_people_evolution = pd.DataFrame(columns = list(dys_df.columns) + ['Date'])
dysgraphia_people_evolution.head()
```

```python
import numpy as np
import datetime

learning_rate = 0.1


current_date = datetime.date(2023, 3, 24)
dates = [current_date + datetime.timedelta(days = 30 * i) for i in range(12)]


for id in list(dys_df.ID):
  person = dys_df[dys_df['ID'] == id]
  age = int(person['Age'].values)
  ages = [age] * 12
  

  X_pred = pd.DataFrame(index = range(12), columns = dysgraphia_people_evolution.columns)
  X_pred['Age'], X_pred['Date'], X_pred['ID'] = ages, dates, [id] * 12
  for f in features:
    X_pred.at[0, f] = person[f].values[0]

  
  for i in range(1, 12):
      
      distances = np.sum(np.square(normal_df[features] - X_pred.iloc[i-1][features]), axis=1)
      
      closest_index = np.argmin(distances)
      
      
      diff = normal_df[features].iloc[closest_index] - X_pred.iloc[i-1][features]
      
      
      X_pred.at[i, tuple(features)] = X_pred.iloc[i-1][features] + learning_rate * diff
  dysgraphia_people_evolution = pd.concat([dysgraphia_people_evolution, X_pred], axis = 0)
```

```python
dysgraphia_people_evolution
```

```python
dysgraphia_people_evolution.to_excel('/content/dysgraphic_people_samples.xlsx', index = None)
```
